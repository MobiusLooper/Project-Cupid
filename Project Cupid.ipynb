They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information.

## - Data Wrangling

````python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import defaultdict
import seaborn as sns
from math import trunc

%matplotlib inline
````

Then we read in our data file, dropping the few lines that do not have have `pid` data, which we need for looking up partner details.

````python
df = pd.read_csv('Speed Dating Data.csv', encoding="ISO-8859-1",thousands=',')
df.dropna(subset=['pid'], inplace = True)
df.reset_index(inplace = True,drop = True)
````

However, we need to allow each row to include a reference to the correspondong partner's row, which we do now. After creating the `partnerindex` function, we call it on the entire data set and create a new column of partner references.

````python
def partnerindex(x):
    return df[(df.iid == int(df.loc[[x]].pid)) & (df.pid == int(df.loc[[x]].iid))].index

for Index,row in df.iterrows():
    df.loc[[Index],'partner_index'] = list(partnerindex(df.index.get_loc(Index)))[0]
````

Now we bucket careers and fields of study into categories since the boolean test for whether exact careers/fields match be ```True``` infrequently. Categorizing careers should add more frequency to ```True``` values.

````python
field_cat_dict = {1: {2,4,5},             # quantitative
                  2: {1,6,7,13,14,15,16}, # qualitative
                  3: {3,8,10,17},         # qualitative and quantitative
                  4: {9,11},              # social
                  5: {12,18}}             # other

def field_cat_function(field_cd):
    for key in field_cat_dict.keys():
        if field_cd in field_cat_dict[key]:
            return key
        
df['field_cat'] = df.field_cd.apply(field_cat_function)
````

````python
career_cat_dict = {1: {4,5},         # quantitative
                   2: {1,3,6,13,16}, # qualitative
                   3: {2,7,8,17},    # qualitative and quantitative
                   4: {9,11,12},     # social
                   5: {10,14,15}}    # other

def career_cat_function(career_c):
    for key in career_cat_dict.keys():
        if career_c in career_cat_dict[key]:
            return key
        
df['career_cat'] = df.career_c.apply(career_cat_function)
````

We also need to create a new dataframe that is ordered by partner index. This way, we can create new comparison features easily in the existing dataset by doing operations on one feature column from each dataframe at a time.

````python
df_partners = df.sort_values('partner_index').reset_index()
````

Given the wrangling done above, creation of new comparison features is simple.

````python
df['int_corr']       = df['int_corr'] # no change
df['samerace']       = df['samerace'] # no change
df['age_diff']       = df.age - df.age_o
df['field_same']     = df.field_cd == df_partners.field_cd
df['field_cat_same'] = df.field_cat == df_partners.field_cat
df['sat_diff']       = df.mn_sat - df_partners.mn_sat
df['income_diff']    = df.income - df_partners.income
df['goal_match']     = df.goal == df_partners.goal
df['date_freq']      = df.date - df_partners.date
df['go_out_freq']    = df.go_out - df_partners.go_out
df['career_same']    = df.career_c == df_partners.career_c
df['career_cat_same']= df.career_cat == df_partners.career_cat
df['opt_diff']       = df.exphappy - df_partners.exphappy
df['expnum_diff']    = df.expnum - df_partners.expnum
df['tuition_diff']   = df.tuition - df_partners.tuition
df['self_score_diff']= (df.attr5_1+df.sinc5_1+df.intel5_1+df.fun5_1+df.amb5_1)-(df_partners.attr5_1+df_partners.sinc5_1+df_partners.intel5_1+df_partners.fun5_1+df_partners.amb5_1)
````

Now we create a trimmed down data set that only consists of the features we are interested in, and the _match_ column. Since each date is accounted for twice, we drop males from the data set without loss of generality, giving us a new dataset that includes each date only once, but contains all the information we want about the pair.

````python
dfNew = df[df['gender'] == 0] # gender being zero denotes females
dfDates = dfNew
new_df_columns = [
    'match',
    'wave',
    'int_corr',
    'samerace',
    'age_diff',
    'field_same',
    'field_cat_same',
    'sat_diff',
    'income_diff',
    'goal_match',
    'date_freq',
    'go_out_freq',
    'career_same',
    'career_cat_same',
    'opt_diff',
    'expnum',
    'tuition_diff',
    'self_score_diff'
]

dfNew = dfNew[new_df_columns].reset_index()
````

For example, given most of our features involve the difference in one comparative metric between females and males, we can see whether our features have any bias. This z-score isn't high (in absolute value terms) enough to meaningfully conclude, using hypothesis testing, that there is a bias for men to be more optimistic, but it is an interesting observation nonetheless that males in this dataset were slightly more optimistic.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import make_scorer
````

We need to make sure we choose the right metric to judge our model's efficacy. To do this, we have to know how many dates end up with a match.

````python
TotalMatches = len(dfNew[dfNew.match == 1])
TotalDates = len(dfNew)

print("Matches to total dates: {0:.2f}%".format(TotalMatches/TotalDates*100))
````

With only one sixth of the data belonging to the `match` class, we have to choose our metric wisely. If we just chose simple accuracy (% of predicted classes matching actual class), we could make a 84% accurate model by simply guessing no match for each date. Our score would be good, but the model would be trivially useless.

We use precision for the positive `match` label as our metric. This is defined as:

$$\frac{\text{Correct match predictions}}{\text{Total match predictions}}$$

The rational behind this is that, for a platform trying to match partner candidates, the objective is to maximise the likelihood that a suggested match will lead to a successful date (with success defined as both parties wishing to pursue the relationship further). The complement to this metric, _recall_, is not as important because we do not need, as a dating platform, to make sure that we recommend all possible matches that will be successful, but rather that the quality of recommendation is good.

Now we create splits of the data for cross validation

````python
# Preparing cross validation training/test set pairs
from sklearn.model_selection import StratifiedKFold
    
X = dfNew.drop(['match','index','wave'],1) 
y = dfNew['match']

# Divide data up into stratified train/test sets
CV = StratifiedKFold(n_splits=4, shuffle = True)
````

We are now able to build and test our first model.

````python
# Support vector classifier with linear kernel
from sklearn.model_selection import cross_val_score

np.random.seed(0)
    
score_LSVC = np.mean(cross_val_score(SVC(), X, y, cv=CV.split(X,y),scoring='precision'))

print('LSVC scores {0:.2f}%'.format(score_LSVC*100))
````

The problem here is that no matches have been predicted due to the sparsity of the `match` label. We can combat this by duplicating dates with `match` labels, so that our new data is made up of 50% matches and 50% dates without a resulting match. Given that dates without matches outnumber matches approximately by five to one, we can clone four copies of the 'no match' data into the dataset to get a fairly well balanced data set.

````python
# clone match rows in dataframe

dfBalanced = dfNew.append([dfNew.loc[dfNew.match == 1]]*4,ignore_index=True)

BalancedDates = len(dfBalanced.loc[dfBalanced.match == 0])/len(dfBalanced)

print("Matches to total dates in balanced data set: {0:.2f}%".format(BalancedDates*100))
````

We now see positive predictions, but have little idea about whether the model is of any use. 
We need to see whether the trained model generalises back to the original mix of classes (approx. 1/6 matches). After all, the model will never know explicitly how many of the couples it looks at are potential matches.

Below, we create our own CV function to accommodate our method of training on balanced data but testing on data with the original mix of classes.

````python
def CV_balanced(model,df,scorer,folds=4):
    """
    Given a valid model, this function first splits the df into k-fold CV sets of training
    and test data, converts the training data into a balanced data set (by class) by cloning
    the rarer class and then trains the model on this data. The model is then tested on data
    with the original distribution of classes and then returns the average performance 
    (according to the specified scorer) and variance of scores over the folds.
    """
    # initialise lists
    train_indices,test_indices = [],[]
    scores = []
    
    # generate CV row indices
    CV = StratifiedKFold(n_splits=folds, shuffle = True)
    for train_index,test_index in CV.split(df.drop(['match'],1),df['match']):
        train_indices.append(train_index)
        test_indices.append(test_index)
     
    # perform k-fold CV
    for fold in range(folds):
        dfTrain = df.loc[train_indices[fold]].reset_index().drop(['level_0'],axis=1)
        dfTest  = df.loc[test_indices[fold]].reset_index().drop(['level_0'],axis=1)
        
        # work out how many clones to add for balanced data
        training_matches = dfTrain.loc[dfTrain.match == 1].reset_index().drop(['level_0'],axis=1)
        num_matches = len(training_matches)
        match_prop = num_matches/len(dfTrain)
        # the below line of code mathematically produces a multiple of training_matches that
        # will ensure a 50/50 mix of matches in the training data
        num_clones_to_add = (1-2*match_prop)/match_prop
        num_clones_trunc = trunc(num_clones_to_add)
        num_clones_rem = num_clones_to_add - num_clones_trunc
        
        # append clones
        clones_to_add_trunc = [training_matches]*num_clones_trunc
        clones_to_add_rem = [training_matches.loc[range(int(num_clones_rem*num_matches))]]
        dfTrain = dfTrain.append(clones_to_add_trunc,ignore_index=True)
        dfTrain = dfTrain.append(clones_to_add_rem,ignore_index=True)
        
        # generate train/test data
        X_train = dfTrain.drop(['match','index','wave'],1) 
        y_train = dfTrain['match']
        X_test = dfTest.drop(['match','index','wave'],1) 
        y_true = dfTest['match']
        
        # run model and record score
        y_pred = model.fit(X_train,y_train).predict(X_test)
        score = scorer(y_true,y_pred,average='binary')
        scores.append(score)
        
        avg_score = np.mean(scores)
        stdv_score = (np.var(scores))**0.5 # useful for fine tuning a model
        
    return avg_score,stdv_score
````

Now we test various models to see which performs best, and take this forward. We also test with a dummy classifier to establish a baseline.

Using either the `stratified` or `uniform` strategies, the dummy classifier scores approximately 16%, which makes sense since this is the proportion of the original data set that belong to the `match` class. Therefore, we need any more sophisicated models to beat that benchmark in order to be valuable.

````python
np.random.seed(0)

model_list = [
    ['Logistic Regression', LogisticRegression()],
    ['AdaBoost', AdaBoostClassifier()],
    ['Decision Tree', DecisionTreeClassifier()],
    ['Linear SVC', LinearSVC()],
    ['SVC', SVC()],
    ['Random Forest', RandomForestClassifier()],
    ['K-Neighbors Classification', KNeighborsClassifier()],
    ['Gaussian Naive Bayes',GaussianNB()]
]

print('{0:>30}    | Model precision | Excess precision over dummy classifier\n'.format('Model type'),'-'*78)

stored_scores = []

for model in model_list:
    model_name = model[0]
    model_score = CV_balanced(model[1],dfNew,precision_score,folds=6)[0]
    stored_scores.append(model_score)
    print('{0:>30}    |     {1:.2f}%      | {2:.2f}%'.format(model_name,model_score*100,model_score*100-16))
````

### Deployment Plan

1. Devise an A/B test, which enables for direct comparison between Tinder’s existing random classifier vs. our predictive model.

2. To do so, we will need to randomly sample a set of users to form our control group (using existing algorithm) and a random set of users to our treatment group (using Cupid predictive model). 

3. By sampling Tinder’s user base at random, we can be sure that the two test groups will be statistically equivalent and so both groups will have the same overall behavioural outcome. 

4. The design of the experiment will be as follows:
    -   Null Hypothesis – Deploying the Cupid predictive model will not increase matches.

    -	Alternative Hypothesis – Deploying the Cupid predictive model will increase matches. 

    -   80% power of test with a 5% level of signficance (this ensures that we do not have to run the test for a long period of time as to achieve above 80% power of test we must increase testing samples signficantly which will lead to higher cost).

5. Run test for 90 days to minimise any disruption on Tinder's platform

6. Once the experiment is complete, we can begin to analyse the data. We can identify the rate of change in matches and Tinder’s analytics platform will also give us some insight as to how user engagement and acquisition was effected using the Cupid predictive model.
